```python
import numpy as np
from matplotlib import transforms
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
import datetime
import time

importances_exp1 = np.load("numpy_objects/experiment_1_importances.npy")
importances_exp2 = np.load("numpy_objects/experiment_2_importances.npy")
importances = [importances_exp1, importances_exp2]

n_features = 11

fig, axes = plt.subplots(1,2)
fig.set_size_inches(25, 10)

for index, ax in enumerate(axes):
    ax.set_xlim([-0.6, n_features-0.4])
    for tick in ax.get_xticklabels():
        tick.set_fontsize(18)
        tick.set_rotation(45)
        tick.set_ha("right")
        tick.set_rotation_mode("anchor")
    for tick in ax.get_yticklabels():
        tick.set_fontsize(18)
    
    importances_with_labels_sorted = importances[index]
    l = len(importances_with_labels_sorted)
    importances_with_labels_sorted = list(importances_with_labels_sorted)[l-n_features:l]
    importances_with_labels_sorted.reverse()
    
    labels = list(map(lambda x: x[0], importances_with_labels_sorted))
    data = list(map(lambda x: float(x[1]), importances_with_labels_sorted))
    bars = ax.bar(labels, data)

    for index, bar in enumerate(bars):
        if ".50" in labels[index]:
            bar.set_color("#333333")
        else:
            bar.set_color("#999999")
plt.show()
```
