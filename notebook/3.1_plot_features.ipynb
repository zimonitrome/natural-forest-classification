```python
import numpy as np
import nfhelpers as nf

sampleAreas = np.load("numpy objects/training_data_set_with_features.npy")
```

```python
# Plot one feature

import matplotlib.pyplot as plt

feature = "BMmean"

positives = list(filter(lambda s: s.classification == 25, sampleAreas))
negatives = list(filter(lambda s: s.classification == 27, sampleAreas))

positives_x_axis = range(0, len(positives))
positives_y_axis = list(map(lambda s: s.features[feature], positives))

negatives_x_axis = range(len(positives), len(positives) + len(negatives))
negatives_y_axis = list(map(lambda s: s.features[feature], negatives))

plt.scatter(positives_x_axis, positives_y_axis, label="positives", color="xkcd:grass green")
plt.scatter(negatives_x_axis, negatives_y_axis, label="negatives", color="red")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel(feature)

plt.show()
```

```python
# Plot two features

import matplotlib.pyplot as plt

feature1 = "H.128.kurt"
feature2 = "H.64.kurt"

positives = list(filter(lambda s: s.classification == 25, sampleAreas))
negatives = list(filter(lambda s: s.classification == 27, sampleAreas))

positives_x_axis = list(map(lambda s: s.features[feature2], positives))
positives_y_axis = list(map(lambda s: s.features[feature1], positives))

negatives_x_axis = list(map(lambda s: s.features[feature2], negatives))
negatives_y_axis = list(map(lambda s: s.features[feature1], negatives))

plt.figure(figsize=(25,15))

plt.scatter(negatives_x_axis, negatives_y_axis, label="negatives", color="red")
plt.scatter(positives_x_axis, positives_y_axis, label="positives", color="xkcd:grass green")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel(feature1)
plt.xlabel(feature2)

plt.show()
```

Plot all features used.

```python
import matplotlib.pyplot as plt

positives = list(filter(lambda s: s.classification == 25, sampleAreas))
negatives = list(filter(lambda s: s.classification == 27, sampleAreas))

col_numb = 4
row_numb = int(len(sampleAreas[0].features) / col_numb) + 1

scatter_pointSize = 20

fig, axarr = plt.subplots(row_numb, col_numb)
fig.set_size_inches(20, 40)
fig.tight_layout()

# Setup multilpe graphs
for  i, feature in enumerate(sampleAreas[0].features):   
    
    positives_x_axis = range(0, len(positives))
    positives_y_axis = list(map(lambda s: s.features[feature], positives))

    negatives_x_axis = range(len(positives), len(positives) + len(negatives))
    negatives_y_axis = list(map(lambda s: s.features[feature], negatives))

    axarr[(int(i/col_numb),i%col_numb)].scatter(positives_x_axis, positives_y_axis, label="positives", color="xkcd:grass green", s=scatter_pointSize)
    axarr[(int(i/col_numb),i%col_numb)].scatter(negatives_x_axis, negatives_y_axis, label="negatives", color="red", s=scatter_pointSize)
    
    axarr[(int(i/col_numb),i%col_numb)].set_title(feature)

plt.show()
```
